```python
import folium
import json
import ipywidgets as widgets
from ipywidgets import interact
from branca.colormap import linear
import numpy as np

# Загружаем данные из файла
with open('migration_probabilities.json', 'r') as file:
    data = json.load(file)

# Все ключи из данных
keys = list(data.keys())

# Функция для отображения карты
def show_map(key):
    # Получаем список точек для выбранного ключа
    points = data[key]

    # Сортируем точки по второй координате (longitude)
    points = sorted(points, key=lambda point: float(point[0].split(',')[1]))

    # Создаем карту, центрируем ее на выбранном ключе
    map_center = [float(key.split(',')[0]), float(key.split(',')[1])]
    m = folium.Map(location=map_center, zoom_start=4)

    # Получаем все значения, чтобы определить максимальное и минимальное значения
    values = [point[1] for point in points]
    min_value = min(values)
    max_value = max(values)

    # Применяем логарифмическое преобразование с масштабированием
    log_values = np.log10(np.array(values) + 1e-100)  # Логарифмируем значения

    # Задаем диапазон цветовой шкалы вручную
    colormap_min = -90  # Минимальное значение логарифмической шкалы
    colormap_max = np.max(log_values)  # Максимальное значение логарифмической шкалы

    # Создаем линейную цветовую шкалу от синего (низкие значения) до красного (высокие значения)
    colormap = linear.RdYlBu_09.scale(colormap_min, colormap_max)

    # Добавляем точки на карту с соответствующими цветами
    for point, log_value in zip(points, log_values):
        coord = point[0].split(',')
        latitude = float(coord[0])
        longitude = float(coord[1])

        # Получаем цвет из шкалы
        color = colormap(log_value)

        folium.CircleMarker(
            location=[latitude, longitude],
            radius=5,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7
        ).add_to(m)

    # Добавляем точку, для которой строится тепловая карта (выбранный ключ)
    selected_lat, selected_lon = map(float, key.split(','))
    folium.Marker(
        location=[selected_lat, selected_lon],
        popup=f"Selected Key: {key}",
        icon=folium.Icon(color='green', icon='star')  # Зеленая звездочка для выделения
    ).add_to(m)

    # Добавляем шкалу на карту
    colormap.caption = 'Log10(Value Scale)'
    colormap.add_to(m)

    # Отображаем карту
    return m

# Создаем выпадающий список для выбора ключа
dropdown = widgets.Dropdown(
    options=keys,
    description='Select Key:',
    disabled=False
)

# Отображаем карту в зависимости от выбора
interact(show_map, key=dropdown)
```

```python
import folium
import json
import ipywidgets as widgets
from ipywidgets import interact
from branca.colormap import linear
import numpy as np

# Загружаем данные из файла
with open('migration_probabilities_fall.json', 'r') as file:
    data = json.load(file)

# Все ключи из данных
keys = list(data.keys())

# Функция для отображения карты
def show_map(key):
    # Получаем список точек для выбранного ключа
    points = data[key]

    # Сортируем точки по второй координате (longitude)
    points = sorted(points, key=lambda point: float(point[0].split(',')[1]))

    # Создаем карту, центрируем ее на выбранном ключе
    map_center = [float(key.split(',')[0]), float(key.split(',')[1])]
    m = folium.Map(location=map_center, zoom_start=4)

    # Получаем все значения
    values = [point[1] for point in points]

    # Линейное масштабирование значений в диапазон [0, 1]
    min_value = min(values)
    max_value = max(values)
    scaled_values = (np.array(values) - min_value) / (max_value - min_value + 1e-10)  # Добавляем 1e-10, чтобы избежать деления на ноль

    # Увеличиваем чувствительность за счет нелинейного преобразования (например, квадратного корня)
    scaled_values = np.sqrt(scaled_values)  # Квадратный корень усиливает различия

    # Создаем линейную цветовую шкалу от синего (низкие значения) до красного (высокие значения)
    colormap = linear.RdYlBu_09.scale(0, 1)

    # Добавляем точки на карту с цветами и размерами, зависящими от значений
    for point, scaled_value, value in zip(points, scaled_values, values):
        coord = point[0].split(',')
        latitude = float(coord[0])
        longitude = float(coord[1])

        # Получаем цвет из шкалы
        color = colormap(scaled_value)

        # Размер маркера зависит от значения (минимальный радиус 2, максимальный 10)
        radius = 2 + 8 * scaled_value

        folium.CircleMarker(
            location=[latitude, longitude],
            radius=radius,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7,
            popup=f"Value: {value:.10f}"  # Добавляем подпись с точным значением
        ).add_to(m)

    # Добавляем точку, для которой строится тепловая карта (выбранный ключ)
    selected_lat, selected_lon = map(float, key.split(','))
    folium.Marker(
        location=[selected_lat, selected_lon],
        popup=f"Selected Key: {key}",
        icon=folium.Icon(color='green', icon='star')  # Зеленая звездочка для выделения
    ).add_to(m)

    # Добавляем шкалу на карту
    colormap.caption = 'Scaled Value Scale'
    colormap.add_to(m)

    # Отображаем карту
    return m

# Создаем выпадающий список для выбора ключа
dropdown = widgets.Dropdown(
    options=keys,
    description='Select Key:',
    disabled=False
)

# Отображаем карту в зависимости от выбора
interact(show_map, key=dropdown)
```
